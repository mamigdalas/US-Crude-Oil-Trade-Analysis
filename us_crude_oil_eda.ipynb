```python
import pandas as pd

# Load the header rows to extract series IDs and descriptive names
# We need rows 0, 1, 2 to correctly identify the true headers and initial data
header_rows_for_imports = pd.read_excel('PET_MOVE_WIMPC_S1_W.xls', sheet_name='Data 1', nrows=3, header=None)

# Extract series IDs (from row 1, index 1)
series_ids = header_rows_for_imports.iloc[1, 1:].tolist() # Skip the first column which is 'Sourcekey'

# Extract descriptive names (from row 2, index 2) to help identify countries
descriptions = header_rows_for_imports.iloc[2, 1:].tolist() # Skip the first column which is 'Date'

# Create a mapping from series ID to a more readable country name
# The first column will be 'Date'
column_names_imports = ['Date']
country_mapping = {}

for i, series_id in enumerate(series_ids):
    description = descriptions[i]
    if 'Weekly U.S. Imports from ' in str(description):
        country_name = description.split('Weekly U.S. Imports from ')[1].split(' of Crude Oil')[0].strip()
        column_name = f'Imports - {country_name}'
        country_mapping[series_id] = column_name
    else: # If it's not a direct country import, use the series ID or a generic name
        column_name = str(series_id)
        country_mapping[series_id] = column_name
    column_names_imports.append(column_name)

# Now, load the actual data. The data itself starts from row 3 (index 2 in the raw file).
# So, we skip 2 rows.
df_imports = pd.read_excel('PET_MOVE_WIMPC_S1_W.xls', sheet_name='Data 1', skiprows=2)

# The loaded df_imports will have the series IDs as column names.
# We need to rename them using our `country_mapping`.
# The first column 'Date' is already correct.
# We'll map the series IDs to the clean names.
df_imports = df_imports.rename(columns={'Date': 'Date'}) # Ensure 'Date' column is explicitly kept if needed

# Dynamically rename other columns based on the country_mapping
# We only want to rename columns that exist in our mapping.
current_cols = df_imports.columns.tolist()
rename_dict = {}
for col in current_cols:
    if col in country_mapping:
        rename_dict[col] = country_mapping[col]
    elif col == 'Date':
        rename_dict[col] = 'Date' # Keep 'Date' as is
    # For any columns not in country_mapping (e.g., Unnamed: 13) we'll leave them as is for now
    # or handle them based on inspection.

df_imports = df_imports.rename(columns=rename_dict)

# Convert 'Date' column to datetime objects
df_imports['Date'] = pd.to_datetime(df_imports['Date'])

# Drop any entirely empty columns that might have come along, for instance Unnamed: 13
df_imports = df_imports.dropna(axis=1, how='all')

print("\n--- Cleaned Imports Data Head ---")
print(df_imports.head())
print("\n--- Cleaned Imports Data Info ---")
print(df_imports.info())
print("\n--- Imports Data Columns ---")
print(df_imports.columns)
```

```python
import pandas as pd

# Define the exports file name
exports_file = 'PET_MOVE_EXPC_A_EP00_EEX_MBBL_M.xls'

# 1. Inspect the raw exports Excel file to identify true start of data and header
# Let's load the first 10 rows from 'Data 1' sheet without header to see its structure
try:
    raw_exports_preview = pd.read_excel(exports_file, sheet_name='Data 1', nrows=10, header=None)
    print("\n--- Raw Exports Data Preview (first 10 rows) ---")
    print(raw_exports_preview)
except FileNotFoundError:
    print(f"Error: '{exports_file}' not found. Please check the file name and ensure it's uploaded.")
except Exception as e:
    print(f"An unexpected error occurred while loading raw exports data preview: {e}")

print("\n" + "="*70 + "\n")

# Based on common EIA patterns for this type of pivot table:
# - Row 0-3 are likely metadata.
# - Row 4 (index 4 in raw Excel) likely contains the "Period-Unit" and then month-year headers (e.g., "Jan-2020", "Feb-2020").
# - The actual data (country names and values) would start from row 5 (index 5 in raw Excel).
# - The first column (index 0) will contain the country names.

try:
    # Read the data, skipping rows until the actual data starts (i.e., skipping 4 rows to get to row 5, which is index 4 in pandas after skiprows)
    # The header=0 means that the first row of the *loaded* data (which was row 5 in Excel) will be used as headers.
    # However, since the columns are dates, it's better to load without a header first, then set the first row as columns.
    # Let's refine this to explicitly get the country names and then the dates.

    # Step A: Load the row that contains the month headers (row 4 of Excel, so skiprows=3)
    # And load the data from where it actually starts (row 5 of Excel, so skiprows=4)
    # We'll use the 'index_col=0' to make the first column (countries) the index for easy transposing.

    # Load the header row which contains the dates (months)
    date_header_row = pd.read_excel(exports_file, sheet_name='Data 1', skiprows=3, nrows=1, header=None)
    # The first element will be 'Period-Unit', then the dates
    date_columns = date_header_row.iloc[0, 1:].tolist() # Skip the first column, take dates

    # Load the main data, skipping the initial header rows, and treating the first column as the index (countries)
    df_exports_raw_data = pd.read_excel(exports_file, sheet_name='Data 1', skiprows=4, header=None)

    # Assign dates to the columns (excluding the first column, which is the country)
    # Ensure the number of date_columns matches the number of data columns after the first country column
    if len(date_columns) == df_exports_raw_data.shape[1] - 1:
        new_cols_for_data = ['Country'] + date_columns
        df_exports_raw_data.columns = new_cols_for_data
    else:
        print("Warning: Mismatch in number of columns and date headers. Proceeding with generic column names for now.")
        # Fallback to generic names if mismatch, we'll need to manually inspect
        df_exports_raw_data.columns = ['Country'] + [f'Col_{i}' for i in range(1, df_exports_raw_data.shape[1])]


    # Set 'Country' as index before transposing
    df_exports = df_exports_raw_data.set_index('Country')

    print("\n--- Exports Data After Initial Load (before transpose) ---")
    print(df_exports.head())
    print("\n--- Exports Data Info (before transpose) ---")
    print(df_exports.info())

    # Transpose the DataFrame
    df_exports_transposed = df_exports.T

    # Convert the index (which are now the dates) to datetime objects
    # Errors='coerce' will turn any unparseable dates into NaT (Not a Time)
    df_exports_transposed.index = pd.to_datetime(df_exports_transposed.index, errors='coerce')

    # Drop any rows where the Date conversion failed (e.g., if there's non-date info that became part of the index)
    df_exports_transposed = df_exports_transposed.dropna(subset=[df_exports_transposed.index.name])

    # Rename the index to 'Date' for clarity
    df_exports_transposed.index.name = 'Date'

    # Clean up column names - some might be summary rows like 'Total All Countries'
    # We will refine this after seeing the head/info
    # For now, let's convert all numeric columns to appropriate type, coercing errors
    for col in df_exports_transposed.columns:
        df_exports_transposed[col] = pd.to_numeric(df_exports_transposed[col], errors='coerce')

    # Drop columns that are entirely NaN after conversion (e.g., if there were empty country rows)
    df_exports_transposed = df_exports_transposed.dropna(axis=1, how='all')

    print("\n--- Final Cleaned Exports Data Head (after transpose) ---")
    print(df_exports_transposed.head())
    print("\n--- Final Cleaned Exports Data Info (after transpose) ---")
    print(df_exports_transposed.info())
    print("\n--- Exports Data Columns ---")
    print(df_exports_transposed.columns.tolist())

except Exception as e:
    print(f"An error occurred while loading exports data: {e}")
```

```python
import pandas as pd

# Define the new exports file name
new_exports_file = 'crude_oil_exports_by_destination.xls'

# 1. Inspect the raw exports Excel file to identify true start of data and header
# Let's load the first 10 rows from 'Data 1' sheet without header to see its structure
try:
    raw_exports_preview_new = pd.read_excel(new_exports_file, sheet_name='Data 1', nrows=10, header=None)
    print("\n--- NEW Raw Exports Data Preview (first 10 rows) ---")
    print(raw_exports_preview_new)
except FileNotFoundError:
    print(f"Error: '{new_exports_file}' not found. Please check the file name and ensure it's uploaded.")
except Exception as e:
    print(f"An unexpected error occurred while loading new raw exports data preview: {e}")

print("\n" + "="*70 + "\n")

# Based on previous experience with EIA "by destination" files and the expected structure:
# - Row 0-3 are typically metadata.
# - Row 4 (index 4 in raw Excel) often contains the "Period-Unit" header in the first column, and then month-year headers (e.g., "Jan-2020", "Feb-2020") starting from the second column.
# - The actual data (country names in the first column, values in subsequent columns) would start from row 5 (index 5 in raw Excel).

try:
    # Read the data, skipping the initial metadata rows.
    # We will set the header to the row that contains the month/year information.
    # If month/year is in row 4 (index 3 if skiprows=3), then header=0 means the first row after skipping is the header.
    df_exports_raw_pivot = pd.read_excel(new_exports_file, sheet_name='Data 1', skiprows=4)

    print("\n--- NEW Exports Data After Initial Load (before transpose) ---")
    print(df_exports_raw_pivot.head())
    print("\n--- NEW Exports Data Info (before transpose) ---")
    print(df_exports_raw_pivot.info())

    # The first column should now be the country names. Let's rename it to 'Destination Country'.
    # Access the first column name using its position, as the actual name might vary (e.g., 'U.S. Exports to...')
    df_exports_raw_pivot = df_exports_raw_pivot.rename(columns={df_exports_raw_pivot.columns[0]: 'Destination Country'})

    # Set 'Destination Country' as the index before transposing
    df_exports_transposed = df_exports_raw_pivot.set_index('Destination Country').T

    # The index is now dates, and columns are countries. Rename the index to 'Date'.
    df_exports_transposed.index.name = 'Date'

    # Convert the index to datetime objects. 'errors='coerce' will turn any unparseable dates into NaT.
    df_exports_transposed.index = pd.to_datetime(df_exports_transposed.index, errors='coerce')

    # Drop any rows (dates) where the conversion failed (e.g., if there were summary rows that became part of the index).
    df_exports_transposed = df_exports_transposed.dropna(subset=[df_exports_transposed.index.name])

    # Clean up column names: remove 'U.S. Exports to ' and ' of Crude Oil (Thousand Barrels)'.
    # Also, some columns might be 'Total all countries' or similar summaries that we might want to filter out.
    cleaned_columns = []
    for col in df_exports_transposed.columns:
        col_str = str(col)
        if 'U.S. Exports to ' in col_str and ' of Crude Oil (Thousand Barrels)' in col_str:
            country_name = col_str.replace('U.S. Exports to ', '').replace(' of Crude Oil (Thousand Barrels)', '').strip()
            cleaned_columns.append(f'Exports - {country_name}')
        elif 'Total all countries' in col_str or 'Total all countries ' in col_str:
            cleaned_columns.append('Exports - Total all countries')
        else:
            cleaned_columns.append(f'Exports - {col_str.strip()}') # Catch-all for other names

    df_exports_transposed.columns = cleaned_columns

    # Convert all data columns to numeric, coercing errors to NaN.
    # This ensures that export volumes are treated as numbers.
    for col in df_exports_transposed.columns:
        df_exports_transposed[col] = pd.to_numeric(df_exports_transposed[col], errors='coerce')

    # Drop any columns (countries) that are entirely NaN after conversion (e.g., if there were empty rows in the original data).
    df_exports_transposed = df_exports_transposed.dropna(axis=1, how='all')

    print("\n--- Final Cleaned Exports Data Head (after transpose) ---")
    print(df_exports_transposed.head())
    print("\n--- Final Cleaned Exports Data Info (after transpose) ---")
    print(df_exports_transposed.info())
    print("\n--- Exports Data Columns ---")
    print(df_exports_transposed.columns.tolist())

except Exception as e:
    print(f"An error occurred while loading and cleaning the new exports data: {e}")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Ensure matplotlib plots are displayed inline in Colab
%matplotlib inline

print("\n--- Analyzing Seasonal Patterns in Total Imports ---")

# Extract the month from the 'Date' column
df_imports['Month'] = df_imports['Date'].dt.month

# Calculate the average total imports for each month
# Exclude the very recent data (e.g., after 2024 if data is incomplete)
# or focus on a more stable period if there were major shifts.
# For now, let's use all available data, but be mindful of the shale boom.
# Let's consider the period AFTER the major import decline has settled a bit for more stable seasonality
# E.g., from 2015 onwards, if that makes sense for import seasonality
# Given the overall decline, just averaging over the full period might obscure recent seasonality.
# Let's average over the entire available period for simplicity first, then discuss.

# Group by month and calculate the mean of 'Imports - Total (Thousand Barrels per Day)'
monthly_avg_imports = df_imports.groupby('Month')['Imports - Total (Thousand Barrels per Day)'].mean()

# Map month numbers to month names for better readability on the plot
month_names = {
    1: 'Jan', 2: 'Feb', 3: 'Mar', 4: 'Apr', 5: 'May', 6: 'Jun',
    7: 'Jul', 8: 'Aug', 9: 'Sep', 10: 'Oct', 11: 'Nov', 12: 'Dec'
}
monthly_avg_imports.index = monthly_avg_imports.index.map(month_names)

plt.figure(figsize=(12, 6))
sns.barplot(x=monthly_avg_imports.index, y=monthly_avg_imports.values, palette='viridis')
plt.title('Average Monthly U.S. Total Crude Oil Imports (Thousand Barrels per Day)')
plt.xlabel('Month')
plt.ylabel('Average Total Imports (Thousand Barrels per Day)')
plt.grid(axis='y', linestyle='--')
plt.tight_layout()
plt.show()

print("\n--- Analyzing Seasonal Patterns in Total Exports ---")

# Extract the month from the 'Date' column
df_exports['Month'] = df_exports['Date'].dt.month

# Calculate the average total exports for each month
# For exports, the policy change in 2015 dramatically altered the pattern.
# Averaging over the *entire* 1920-present period would heavily skew early data.
# It's more meaningful to analyze seasonality *after* the export ban was lifted.
# So, let's filter df_exports for data from 2016 onwards for seasonality analysis.
df_exports_post_ban = df_exports[df_exports['Date'].dt.year >= 2016].copy()

monthly_avg_exports = df_exports_post_ban.groupby('Month')['Exports - Total (Thousand Barrels)'].mean()
monthly_avg_exports.index = monthly_avg_exports.index.map(month_names) # Use the same month_names map

plt.figure(figsize=(12, 6))
sns.barplot(x=monthly_avg_exports.index, y=monthly_avg_exports.values, palette='magma')
plt.title('Average Monthly U.S. Total Crude Oil Exports (Thousand Barrels) (Post-2015)')
plt.xlabel('Month')
plt.ylabel('Average Total Exports (Thousand Barrels)')
plt.grid(axis='y', linestyle='--')
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import plotly.express as px
import plotly.graph_objects as go

# Ensure df_imports and df_exports are loaded and cleaned as per previous steps

print("--- Creating Interactive Time Series Plots ---")

# --- Interactive Total Imports Plot ---
fig_imports_total = px.line(
    df_imports,
    x='Date',
    y='Imports - Total (Thousand Barrels per Day)',
    title='Weekly U.S. Total Crude Oil Imports Over Time (Interactive)',
    labels={
        'Imports - Total (Thousand Barrels per Day)': 'Total Imports (Thousand Barrels per Day)',
        'Date': 'Date'
    }
)
fig_imports_total.update_layout(hovermode="x unified") # Shows all y-values for a given x-date
fig_imports_total.show()


# --- Interactive Total Exports Plot ---
# For exports, remember to filter for post-2015 for meaningful trends
df_exports_post_2015 = df_exports[df_exports['Date'].dt.year >= 2015].copy()

fig_exports_total = px.line(
    df_exports_post_2015,
    x='Date',
    y='Exports - Total (Thousand Barrels)',
    title='Monthly U.S. Total Crude Oil Exports Over Time (Post-2015, Interactive)',
    labels={
        'Exports - Total (Thousand Barrels)': 'Total Exports (Thousand Barrels)',
        'Date': 'Date'
    }
)
fig_exports_total.update_layout(hovermode="x unified")
fig_exports_total.show()

# --- Interactive Imports vs Exports Plot (Common Range) ---
# Re-create the monthly-aggregated data for comparison from previous EDA steps
# For df_imports, we have weekly data in 'Thousand Barrels per Day'.
# To get 'Thousand Barrels per Month':
# 1. Calculate weekly total barrels: daily average * 7 days/week
# 2. Resample to monthly, summing up weekly totals within each month.
imports_cols = [col for col in df_imports.columns if col.startswith('Imports - ') and 'Total' not in col]
df_imports_monthly_agg_for_plot = df_imports.copy()
df_imports_monthly_agg_for_plot['Weekly Total Barrels'] = df_imports_monthly_agg_for_plot[imports_cols].sum(axis=1) * 7

df_imports_resampled_monthly_for_plot = df_imports_monthly_agg_for_plot.set_index('Date')['Weekly Total Barrels'].resample('MS').sum().reset_index()
df_imports_resampled_monthly_for_plot = df_imports_resampled_monthly_for_plot.rename(columns={'Weekly Total Barrels': 'Imports - Total (Thousand Barrels per Month)'})


# For Exports, the data is already monthly total barrels.
df_exports_for_comparison_plot = df_exports[['Date', 'Exports - Total (Thousand Barrels)']].copy()
# Ensure 'Date' for exports is also at the start of the month for alignment
df_exports_for_comparison_plot['Date'] = df_exports_for_comparison_plot['Date'].dt.to_period('M').dt.to_timestamp()
df_exports_for_comparison_plot = df_exports_for_comparison_plot.rename(columns={'Exports - Total (Thousand Barrels)': 'Exports - Total (Thousand Barrels per Month)'})

# Find common date range
min_date_common = max(df_imports_resampled_monthly_for_plot['Date'].min(), df_exports_for_comparison_plot['Date'].min())
max_date_common = min(df_imports_resampled_monthly_for_plot['Date'].max(), df_exports_for_comparison_plot['Date'].max())

df_imports_common_range_plot = df_imports_resampled_monthly_for_plot[(df_imports_resampled_monthly_for_plot['Date'] >= min_date_common) & (df_imports_resampled_monthly_for_plot['Date'] <= max_date_common)]
df_exports_common_range_plot = df_exports_for_comparison_plot[(df_exports_for_comparison_plot['Date'] >= min_date_common) & (df_exports_for_comparison_plot['Date'] <= max_date_common)]

# Create a combined DataFrame for plotting
df_combined_trade = pd.merge(
    df_imports_common_range_plot[['Date', 'Imports - Total (Thousand Barrels per Month)']],
    df_exports_common_range_plot[['Date', 'Exports - Total (Thousand Barrels per Month)']],
    on='Date',
    how='inner'
)

fig_combined_trade = px.line(
    df_combined_trade,
    x='Date',
    y=['Imports - Total (Thousand Barrels per Month)', 'Exports - Total (Thousand Barrels per Month)'],
    title='U.S. Crude Oil Imports vs. Exports Over Common Time Range (Monthly Total, Interactive)',
    labels={
        'value': 'Volume (Thousand Barrels per Month)',
        'variable': 'Trade Type',
        'Date': 'Date'
    }
)
fig_combined_trade.update_layout(hovermode="x unified")
fig_combined_trade.show()
```

```python
import pandas as pd
import plotly.express as px
import dash
from dash import dcc
from dash import html
from dash.dependencies import Input, Output
import warnings

# Suppress specific future warnings from pandas or other libraries that might appear due to chained assignments or similar
warnings.simplefilter(action='ignore', category=FutureWarning)

# --- Data Preparation for Dash App ---
# We need to ensure the dataframes are ready and 'Total' columns exist

# Ensure 'Imports - Total (Thousand Barrels per Day)' exists
if 'Imports - Total (Thousand Barrels per Day)' not in df_imports.columns:
    imports_cols = [col for col in df_imports.columns if col.startswith('Imports - ') and 'Total' not in col]
    df_imports['Imports - Total (Thousand Barrels per Day)'] = df_imports[imports_cols].sum(axis=1)

# Ensure 'Exports - Total (Thousand Barrels)' exists (it should from previous steps)

# Prepare data for top partners
# Imports
import_country_cols_dash = [col for col in df_imports.columns if col.startswith('Imports - ') and 'Total' not in col]
total_imports_by_country_dash = df_imports[import_country_cols_dash].sum().sort_values(ascending=False)
total_imports_by_country_dash.index = total_imports_by_country_dash.index.str.replace('Imports - ', '')
top_10_imports_dash = total_imports_by_country_dash.head(10).reset_index()
top_10_imports_dash.columns = ['Country', 'Total Volume']

# Exports (post-2015 for relevance)
export_country_cols_dash = [col for col in df_exports.columns if col.startswith('Exports - ') and 'Total' not in col]
df_exports_post_2015_dash = df_exports[df_exports['Date'].dt.year >= 2015].copy()
total_exports_by_country_dash = df_exports_post_2015_dash[export_country_cols_dash].sum().sort_values(ascending=False)
total_exports_by_country_dash.index = total_exports_by_country_dash.index.str.replace('Exports - ', '')
top_10_exports_dash = total_exports_by_country_dash.head(10).reset_index()
top_10_exports_dash.columns = ['Country', 'Total Volume']


# Initialize the Dash app
# We'll stick with dash.Dash directly for robustness in Colab, as jupyter_dash sometimes causes issues.
app = dash.Dash(__name__)

app.layout = html.Div([
    html.H1("U.S. Crude Oil Trade Analysis Dashboard"),

    html.Div([
        html.Label("Select Trade Type:"),
        dcc.Dropdown(
            id='trade-type-dropdown',
            options=[
                {'label': 'Imports', 'value': 'Imports'},
                {'label': 'Exports', 'value': 'Exports'}
            ],
            value='Imports', # Default value
            clearable=False
        ),
    ]),

    html.Hr(), # Horizontal line for separation

    html.Div(id='time-series-plot-container'), # Container for the time series plot
    html.Div(id='top-partners-plot-container')  # Container for the top partners bar chart
])

# --- Callbacks ---
@app.callback(
    [Output('time-series-plot-container', 'children'),
     Output('top-partners-plot-container', 'children')],
    [Input('trade-type-dropdown', 'value')]
)
def update_graphs(selected_trade_type):
    if selected_trade_type == 'Imports':
        # Time Series Plot for Imports
        fig_time_series = px.line(
            df_imports,
            x='Date',
            y='Imports - Total (Thousand Barrels per Day)',
            title='Weekly U.S. Total Crude Oil Imports Over Time',
            labels={
                'Imports - Total (Thousand Barrels per Day)': 'Total Imports (Thousand Barrels per Day)',
                'Date': 'Date'
            }
        )
        fig_time_series.update_layout(hovermode="x unified")

        # Top Partners Plot for Imports
        fig_top_partners = px.bar(
            top_10_imports_dash,
            x='Total Volume',
            y='Country',
            orientation='h',
            title='Top 10 U.S. Crude Oil Import Partners (Total Volume)',
            labels={
                'Total Volume': 'Total Volume (Thousand Barrels per Day)',
                'Country': 'Country'
            },
            color='Total Volume',
            color_continuous_scale=px.colors.sequential.Viridis
        )
        fig_top_partners.update_yaxes(categoryarray=top_10_imports_dash['Country'].tolist(), categoryorder='array')

    elif selected_trade_type == 'Exports':
        # Time Series Plot for Exports (post-2015)
        fig_time_series = px.line(
            df_exports_post_2015_dash,
            x='Date',
            y='Exports - Total (Thousand Barrels)',
            title='Monthly U.S. Total Crude Oil Exports Over Time (Post-2015)',
            labels={
                'Exports - Total (Thousand Barrels)': 'Total Exports (Thousand Barrels)',
                'Date': 'Date'
            }
        )
        fig_time_series.update_layout(hovermode="x unified")

        # Top Partners Plot for Exports (post-2015)
        fig_top_partners = px.bar(
            top_10_exports_dash,
            x='Total Volume',
            y='Country',
            orientation='h',
            title='Top 10 U.S. Crude Oil Export Partners (Total Volume, Post-2015)',
            labels={
                'Total Volume': 'Total Volume (Thousand Barrels)',
                'Country': 'Country'
            },
            color='Total Volume',
            color_continuous_scale=px.colors.sequential.Magma
        )
        fig_top_partners.update_yaxes(categoryarray=top_10_exports_dash['Country'].tolist(), categoryorder='array')

    return dcc.Graph(figure=fig_time_series), dcc.Graph(figure=fig_top_partners)

# Corrected call for running the server in Colab
if __name__ == '__main__':
    # Use mode='inline' with the standard dash.Dash app
    # This might require a newer version of dash or specific Colab setup.
    # If it fails, the 'external' mode is the most reliable fallback.
    try:
        app.run(mode='inline', debug=True)
    except Exception as e:
        print(f"Inline mode failed: {e}. Trying external mode...")
        # Fallback to external mode if inline fails
        app.run(mode='external', debug=True, port=8050)
        print("\nIf the inline app doesn't load, please click the link below (might take a moment):")
        print("http://127.0.0.1:8050/") # This is the usual default external link
```
